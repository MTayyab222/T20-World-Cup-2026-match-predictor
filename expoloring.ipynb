```python
from sklearn.preprocessing import LabelEncoder
import pickle
```

```python
# Load both CSV files
dataset = pd.read_csv('Match_dataset.csv')
dataset2 = pd.read_csv('match_results.csv')

# Merge them
merged = dataset.merge(dataset2, on='Match_ID')

print(f"✅ Data loaded successfully!")
print(f"   Match_dataset: {dataset.shape}")
print(f"   match_results: {dataset2.shape}")
print(f"   Merged: {merged.shape}")
```

```python
# All text columns that need encoding
columns_to_encode = [
    'Team_A',
    'Team_B',
    'Venue',
    'Stage',
    'Pitch_Type',
    'Toss_Winner',
    'Toss_Decision'
]

print(f"✅ Will encode {len(columns_to_encode)} columns:")
for i, col in enumerate(columns_to_encode, 1):
    unique = merged[col].nunique()
    print(f"   {i}. {col:20} ({unique} unique values)")
```

```python
# Create copy
df_encoded = merged.copy()

# Dictionary to store encoders
label_encoders = {}

# Encode each column
for col in columns_to_encode:
    # Create encoder
    le = LabelEncoder()
    
    # Encode column
    df_encoded[col + '_encoded'] = le.fit_transform(merged[col])
    
    # Save encoder
    label_encoders[col] = le
    
    # Show example
    original = merged[col].iloc[0]
    encoded = df_encoded[col + '_encoded'].iloc[0]
    print(f"✅ {col:20} : '{original}' → {encoded}")

print(f"\n🎉 All {len(columns_to_encode)} columns encoded!")
```

```python
# Toss Winner Binary
df_encoded['Toss_Winner_Binary'] = merged['Toss_Winner'].map({
    'Team_A': 1,
    'Team_B': 0
})

# Toss Decision Binary
df_encoded['Toss_Decision_Binary'] = merged['Toss_Decision'].map({
    'Bat': 0,
    'Field': 1
})

# Winner Binary
df_encoded['Winner_Binary'] = merged['Winner'].map({
    'Team_A': 1,
    'Team_B': 0
})

print("✅ Binary columns created:")
print("   • Toss_Winner_Binary (1=Team_A, 0=Team_B)")
print("   • Toss_Decision_Binary (0=Bat, 1=Field)")
print("   • Winner_Binary (1=Team_A, 0=Team_B)")
```

```python
# Save encoders
with open('label_encoders.pkl', 'wb') as f:
    pickle.dump(label_encoders, f)

print("✅ Encoders saved to: label_encoders.pkl")
```

```python
# Save encoded dataset
df_encoded.to_csv('data_encoded.csv', index=False)

print("✅ Encoded data saved to: data_encoded.csv")
print(f"   Shape: {df_encoded.shape}")
print(f"   Columns: {df_encoded.shape[1]}")
```

```python
# Show first match - Before and After
print("🔴 BEFORE (Original):")
print("-" * 60)
for col in columns_to_encode[:3]:
    print(f"   {col:15} : {merged[col].iloc[0]}")

print("\n🟢 AFTER (Encoded):")
print("-" * 60)
for col in columns_to_encode[:3]:
    original = merged[col].iloc[0]
    encoded = df_encoded[col + '_encoded'].iloc[0]
    print(f"   {col:15} : {original} → {encoded}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np

print("✅ ML libraries imported!")
```

```python
# Features (Input to model)
feature_columns = [
    'Team_A_encoded',
    'Team_B_encoded',
    'Team_A_Ranking',
    'Team_B_Ranking',
    'Team_A_Form',
    'Team_B_Form',
    'HeadToHead_A_Wins',
    'HeadToHead_B_Wins',
    'Venue_encoded',
    'Venue_HomeAdvantage_A',
    'Venue_HomeAdvantage_B',
    'Pitch_Type_encoded',
    'Avg_T20_Score_Venue',
    'Toss_Winner_Binary',
    'Toss_Decision_encoded',
    'Team_A_Tech_Index',
    'Team_B_Tech_Index',
    'Stage_encoded'
]

# X = Features (what model will learn from)
X = df_encoded[feature_columns]

# y = Target (what model will predict)
y = df_encoded['Winner_Binary']

print(f"✅ Features selected!")
print(f"   X shape: {X.shape} (600 matches × {len(feature_columns)} features)")
print(f"   y shape: {y.shape} (600 winners)")
print(f"\n📊 Winner distribution:")
print(f"   Team_A wins: {y.sum()} ({y.sum()/len(y)*100:.1f}%)")
print(f"   Team_B wins: {len(y)-y.sum()} ({(len(y)-y.sum())/len(y)*100:.1f}%)")
```

```python
# Split data: 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 20% for testing
    random_state=42     # Reproducible results
)

print("✅ Data split complete!")
print(f"\n📊 Training Set:")
print(f"   Samples: {len(X_train)} (80%)")
print(f"   Features: {X_train.shape[1]}")

print(f"\n📊 Testing Set:")
print(f"   Samples: {len(X_test)} (20%)")
print(f"   Features: {X_test.shape[1]}")
```

```python
# Initialize Random Forest
model = RandomForestClassifier(
    n_estimators=100,      # 100 decision trees
    max_depth=10,          # Maximum tree depth
    min_samples_split=5,   # Minimum samples to split
    random_state=42        # Reproducible results
)

print("✅ Random Forest model created!")
print(f"\n⚙️ Model Parameters:")
print(f"   Number of trees: 100")
print(f"   Max depth: 10")
print(f"   Min samples split: 5")
```

```python
print("🎓 Training started...")

# Train the model
model.fit(X_train, y_train)

print("✅ Training complete!")
print("   Model has learned patterns from 480 matches")
print("   Ready to make predictions! 🚀")
```

```python
# Predict on training data
y_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)

# Predict on testing data
y_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("✅ Predictions complete!")
print(f"\n📊 Results:")
print(f"   Training Accuracy: {train_accuracy*100:.2f}%")
print(f"   Testing Accuracy: {test_accuracy*100:.2f}%")
print(f"   Gap: {(train_accuracy - test_accuracy)*100:.2f}%")

if test_accuracy >= 0.80:
    print("\n🌟 EXCELLENT! Model is highly accurate!")
elif test_accuracy >= 0.70:
    print("\n✅ GOOD! Model performs well!")
else:
    print("\n⚠️ Needs improvement")
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_test_pred)

print("📊 Confusion Matrix:")
print("\n          Predicted")
print("          Team_B  Team_A")
print(f"Actual Team_B  {cm[0][0]:3}     {cm[0][1]:3}")
print(f"       Team_A  {cm[1][0]:3}     {cm[1][1]:3}")

print(f"\n💡 Interpretation:")
print(f"   ✅ Correct Team_B predictions: {cm[0][0]}")
print(f"   ✅ Correct Team_A predictions: {cm[1][1]}")
print(f"   ❌ Wrong predictions: {cm[0][1] + cm[1][0]}")
print(f"   Total accuracy: {(cm[0][0] + cm[1][1])/cm.sum()*100:.2f}%")
```

```python
# Get feature importance
feature_importance = pd.DataFrame({
    'Feature': feature_columns,
    'Importance': model.feature_importances_
}).sort_values('Importance', ascending=False)

print("🎯 Top 10 Most Important Features:")
print("=" * 60)
for i, row in feature_importance.head(10).iterrows():
    print(f"{row['Feature']:30} → {row['Importance']:.4f}")

print("\n💡 Higher value = More important for predictions")
```

```python
# Save trained model
with open('cricket_model.pkl', 'wb') as f:
    pickle.dump(model, f)

# Save feature columns list
with open('feature_columns.pkl', 'wb') as f:
    pickle.dump(feature_columns, f)

print("✅ Model saved successfully!")
print("\n📁 Files created:")
print("   • cricket_model.pkl (trained model)")
print("   • feature_columns.pkl (feature list)")
print("   • label_encoders.pkl (from preprocessing)")
```
